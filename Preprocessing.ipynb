```python
import os
from shutil import copyfile
from random import shuffle
```

```python
caltech_dir = '/scratch/hle/101_ObjectCategories'
list_dirs = os.listdir(caltech_dir)
list_dirs.sort()

train_dir = '/scratch/hle/caltech_101/train'
valid_dir = '/scratch/hle/caltech_101/valid'
test_dir = '/scratch/hle/caltech_101/test'

for index, label in enumerate(list_dirs):
    if not os.path.exists(os.path.join(train_dir, label)):
        os.makedirs(os.path.join(train_dir, label))
    if not os.path.exists(os.path.join(valid_dir, label)):
        os.makedirs(os.path.join(valid_dir, label))
    if not os.path.exists(os.path.join(test_dir, label)):
        os.makedirs(os.path.join(test_dir, label))
        
    label_dir = os.path.join(caltech_dir, label)
    file_names = os.listdir(label_dir)
    shuffle(file_names)
    num_files = len(file_names)
    train_files = file_names[:num_files//2]
    valid_files = file_names [num_files//2:(num_files//2) + (num_files//4)]
    test_files = file_names [(num_files//2) + (num_files//4):]
    for f in train_files:
        img_path = os.path.join(label_dir, f)
        copyfile(img_path, os.path.join(train_dir, label, f))
    for f in valid_files:
        img_path = os.path.join(label_dir, f)
        copyfile(img_path, os.path.join(valid_dir, label, f))
    for f in test_files:
        img_path = os.path.join(label_dir, f)
        copyfile(img_path, os.path.join(test_dir, label, f))
```

```python
for index, label in enumerate(list_dirs):
    label_dir = os.path.join(caltech_dir, label)
    total_num_files = len(os.listdir(label_dir))
    
    train_class_dir = os.path.join(train_dir, label)
    num_train_files = len(os.listdir(train_class_dir))
    valid_class_dir = os.path.join(valid_dir, label)
    num_valid_files = len(os.listdir(valid_class_dir))
    test_class_dir = os.path.join(test_dir, label)
    num_test_files = len(os.listdir(test_class_dir))
    print("%s total: %s \t train: %s \t valid: %s \t test: %s" % (label, total_num_files, num_train_files, \
                                                                num_valid_files, num_test_files))
    
```
